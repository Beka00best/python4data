```python
import pandas as pd
from numpy.random import exponential
```

```python
df = pd.DataFrame({'x': range(20), 'y': exponential(10, 20)})
```

```python
import matplotlib.pyplot as plt
```

```python
data1 = exponential(5, 20)
data2 = exponential(6, 20)
```

```python
# plt.figure(figsize=(15, 15))
plt.plot(data1, label='1')
plt.scatter(range(len(data1)), data2, color='red', label='2')
plt.title('Результаты', fontdict={'fontsize': 20})
plt.xlabel('Номер попытки')
plt.ylabel('Результат')
plt.legend()
plt.xticks(range(0, 20, 4))
```

```python
fig, axes = plt.subplots()
axes.plot(exponential(5, 20))
axes.set_title('Graf')
fig, axes = plt.subplots(nrows=2, ncols=3)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 5))
for row, row_exes in enumerate(axes):
    for column, ax in enumerate(row_exes):
        ax.plot(exponential(column, 20))
        ax.set_title('holst kolonka {} stroka {}'.format(column+1, row+1))
fig.tight_layout()
```
